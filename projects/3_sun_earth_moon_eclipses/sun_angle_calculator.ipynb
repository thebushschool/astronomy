# Earth Seasons and Sun Angle Calculations for Seattle and other cities for a given date 

## TASK 1: Let's store the Lattitude and Longitude of Seattle below in decimal degrees for computing
**Seattle Lattitude: 47 36' 28.8468" N**
**Seattle Longitude: 122 20' 6.6012" W**
- Note: North Latitudes are considered Positive and South Negative
- Note: East Longitudes are considered Positive and West Negative

```python
seattle_lattitude = 47 + 36/60 + 28.8468/(60*60)
out = f'Seattle Lattitude = {seattle_lattitude:3.4f} degrees'
print(out)
```

```python
# FIX THESE BELOW!!
seattle_longitude = -1*(122 + 20/60 + 6.6012/(60*60))
out = f'Seattle Longitude = {seattle_longitude:3.4f} degrees'
print(out)
```

## TASK 2: Let's calculate the altitude of the Sun above the horizon in Seattle (which is not on the Tropic of Cancer) at summer solstice June 21st
1. To do this we first need ***Seattle's lattitude which is +47.61 deg***
1. We then need the ***lattitude of the sub-solar point*** on Earth on June 21st. In other words, this is the lattitude at which the Sun is directly overhead.
1. At Summer Solstice, Sun is directly overhead any city on the Tropic of Cancer like Syenne! in the Eratosthenes workshop! At other dates, we can refer to the Analemma chart below. For now, let's proceed with Summer Solstice on Jun 21st for this calculation. 
1. The ***lattitude of the sub-solar point is +23.5 deg***
1. We then find the ***angular separation or angular distance between these two lattitudes***
1. Finally the ***Sun Altitude*** or Sun Elevation above the horizon on Jun 21st is calculated to be ***90 - angular distance*** calculated above

```python
# At summer solstice, Sun is directly overhead any city on the Tropic of Cancer (like Syenne! in the Eratosthenes workshop!)
sub_solar_point = 23.5
sun_altitude_seattle = 90 - (seattle_lattitude - sub_solar_point)
out = f'Sun Altitude Seattle at Summer Solstice (June 21st) = {sun_altitude_seattle:.2f} deg'
print(out)
```

### TASK 3: Using the [Seasons Simulator linked here](https://astro.unl.edu/naap/motion1/animations/seasons_ecliptic.html) snapshot a picture of the summer solstice in the Northern hemisphere. Validate the computer Sun Altitude with your calculation above within 5% accuracy.

```python
sun_altitide_simulator = 65.7 #deg
pct_error = (sun_altitude_seattle - sun_altitide_simulator)/sun_altitide_simulator*100
print(f'Percent Error in calculating Sun Altitiude = {pct_error:.2f}%')
```
